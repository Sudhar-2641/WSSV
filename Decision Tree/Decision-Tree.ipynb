## Model Prediction:

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Example dataset (replace with your actual data)
data = np.random.rand(100, 675)  # 100 samples, 10 features
labels = np.random.randint(0, 2, 100)  # 100 binary labels

# Split the data
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

# Initialize and train the model
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)

# Predict on the test data
y_pred = decision_tree_model.predict(X_test)

# Check shapes
print("y_test shape:", y_test.shape)
print("y_pred shape:", y_pred.shape)

# Compute accuracy score
score = accuracy_score(y_test, y_pred)
print("Accuracy Score:", score)
```

```python
# Make sure y_test corresponds to the X_test data
assert len(y_test) == len(y_pred), "Mismatch between number of test samples and predictions"
```

#### Cross Validation

```python
from sklearn.model_selection import train_test_split

# Split the data
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

# Initialize and fit the model
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)

# Perform cross-validation
cross_val_scores = cross_val_score(decision_tree_model, X_test, y_test, cv=5)

# Print the cross-validation scores
print("Cross-Validation Scores:", cross_val_scores)
print("Mean Cross-Validation Score:", cross_val_scores.mean())
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Cross-validation scores
cross_val_scores = np.array( [0.25, 0.75, 0.5,  0,   0.5 ])
mean_accuracy = cross_val_scores.mean()

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(np.arange(len(cross_val_scores)), cross_val_scores, color='lightblue', label='Fold Accuracy')

# Add a horizontal line for the mean accuracy
plt.axhline(mean_accuracy, color='red', linestyle='--', label=f'Mean Accuracy: {mean_accuracy:.2f}')

# Adding labels, title, and legend
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.title('Cross-Validation Accuracy Scores')
plt.xticks(np.arange(len(cross_val_scores)), [f'Fold {i+1}' for i in range(len(cross_val_scores))])
plt.legend()

# Adding value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}', va='bottom', ha='center')

plt.tight_layout()
plt.show()
```

### Hyperparameter Tunning : GridSearch CV:

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Define the Decision Tree classifier
classifier = DecisionTreeClassifier()

# Define the parameter grid for GridSearchCV
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Setup GridSearchCV
grid_search = GridSearchCV(classifier, param_grid, cv=5, verbose=1)

# Fit GridSearchCV with the training data
grid_search.fit(X_train, y_train)

# Output the best parameters and best score
print("Best parameters found:", grid_search.best_params_)
print("Best cross-validation score:", grid_search.best_score_)
```

```python
best_estimator = grid_search.best_estimator_
best_estimator
```
