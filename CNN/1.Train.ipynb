```python
#connecting dataset with code
import Augmentor
p = Augmentor.Pipeline("/content/drive/My Drive/Major_Project/Image_dataset/train/healthy")
```

```python
#performing rotations on images
p.rotate(probability=0.7, max_left_rotation=10, max_right_rotation=10)
```

```python
#performing zoom-in zoom-out on images
p.zoom(probability=0.3, min_factor=1.1, max_factor=1.6)
```

```python
#taking 500 images as new training dataset
#PS- 500 was just an example I took, we can increase it according to our necssities

training_augmented = p.sample(500)
```
