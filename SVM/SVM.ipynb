## Read the image from the image folder and convert them into matrix it into the Li

```python
input_dir = "/home/sudharsan/Downloads/Image_dataset"
categories = ['healthy','wssv']
data = []
labels = []
for category_idx, category in enumerate(categories):
    for file in os.listdir(os.path.join(input_dir,category)):
        img_path = os.path.join(input_dir,category,file)
        print(img_path)
        img = imread(img_path)
        img = resize(img,(15,15))
        data.append(img.flatten())
        labels.append(category_idx)
```

```python
data[1]
```

```python
labels[1]
```

## List of matrix converted into numpy array:

```python
data = np.asarray(data)
labels = np.asarray(labels)
```

```python
data.shape
```

```python
labels
```

## split the data into training and testing 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(data,labels,test_size = 0.2,shuffle =True,stratify=labels)
```

## Build SVM Model for classification

```python
from sklearn.svm import SVC
svm_model = SVC()
```

## Model training 

```python
svm_model.fit(x_train,y_train)
```

## Model Prediction:

```python
y_pred  = svm_model.predict(x_test)
y_pred
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score 
score = accuracy_score(y_test,y_pred)
score
```

#### Cross Validation

```python
from sklearn.model_selection import cross_val_score
cross_val_score = cross_val_score(svm_model,data,labels,cv = 5)
cross_val_score
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Cross-validation scores
cross_val_scores = np.array([0.83870968, 0.77419355, 0.87096774, 0.77419355, 0.8])
mean_accuracy = cross_val_scores.mean()

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(np.arange(len(cross_val_scores)), cross_val_scores, color='lightblue', label='Fold Accuracy')

# Add a horizontal line for the mean accuracy
plt.axhline(mean_accuracy, color='red', linestyle='--', label=f'Mean Accuracy: {mean_accuracy:.2f}')

# Adding labels, title, and legend
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.title('Cross-Validation Accuracy Scores')
plt.xticks(np.arange(len(cross_val_scores)), [f'Fold {i+1}' for i in range(len(cross_val_scores))])
plt.legend()

# Adding value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}', va='bottom', ha='center')

plt.tight_layout()
plt.show()
```
